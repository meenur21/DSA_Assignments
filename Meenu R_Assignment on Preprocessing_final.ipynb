#### Handling Outliers

    To check outliers using visualization in column 'Fare

```python
import seaborn as sns
sns.distplot(titanic['Fare'])
```

```python
import matplotlib.pyplot as plt
sns.boxplot(x=titanic['Fare'])
plt.title('Boxplot of Fare')
plt.xlabel('Fare')
plt.show()
```

    using Z-Score method

```python
upper_limit = titanic['Fare'].mean()+3*titanic['Fare'].std()
lower_limit = titanic['Fare'].mean()-3*titanic['Fare'].std()
print('Upper limit=',upper_limit)
print('Lower limit=', lower_limit)
```

```python
#now want to find the outliers
titanic.loc[(titanic['Fare']>upper_limit) | (titanic['Fare']<lower_limit)]
```

```python
#changing the values of outliers by capping
df = titanic.copy()
df.loc[(df['Fare']>=upper_limit),'Fare']=upper_limit
df.loc[(df['Fare']<=lower_limit),'Fare']=lower_limit
```

```python
# Visualizing 'Fare' column after handling outliers
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['Fare'])
plt.title('Boxplot of Fare (Outliers Handled)')
plt.xlabel('Fare')
plt.show()
```

    to check in column 'Parch'

```python
sns.distplot(titanic['Parch'])
```

```python
sns.boxplot(x=titanic['Parch'])
plt.title('Boxplot of Parch')
plt.xlabel('Parch')
plt.show()
```

```python
# Z-Score method for 'Parch' column
upper_limit_parch = titanic['Parch'].mean() + 3 * titanic['Parch'].std()
lower_limit_parch = titanic['Parch'].mean() - 3 * titanic['Parch'].std()

# Finding outliers in 'Parch' column
outliers_parch = titanic.loc[(titanic['Parch'] > upper_limit_parch) | (titanic['Parch'] < lower_limit_parch)]

# Handling outliers in 'Parch' column (capping outliers)
df = titanic.copy()
df.loc[(df['Parch'] >= upper_limit_parch), 'Parch'] = upper_limit_parch
df.loc[(df['Parch'] <= lower_limit_parch), 'Parch'] = lower_limit_parch

# Visualizing 'Parch' column after handling outliers
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['Parch'])
plt.title('Boxplot of Parch (Outliers Handled)')
plt.xlabel('Parch')
plt.show()
```

```python
# Visualizing 'Age' column distribution
plt.figure(figsize=(6, 4))
sns.histplot(titanic['Age'], kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()

# Boxplot for 'Age' column
plt.figure(figsize=(6, 4))
sns.boxplot(x=titanic['Age'])
plt.title('Boxplot of Age')
plt.xlabel('Age')
plt.show()

# Z-Score method for 'Age' column
upper_limit_age = titanic['Age'].mean() + 3 * titanic['Age'].std()
lower_limit_age = titanic['Age'].mean() - 3 * titanic['Age'].std()

# Finding outliers in 'Age' column
outliers_age = titanic.loc[(titanic['Age'] > upper_limit_age) | (titanic['Age'] < lower_limit_age)]

# Handling outliers in 'Age' column (capping outliers)
df = titanic.copy()
df.loc[(df['Age'] >= upper_limit_age), 'Age'] = upper_limit_age
df.loc[(df['Age'] <= lower_limit_age), 'Age'] = lower_limit_age

# Visualizing 'Age' column after handling outliers
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['Age'])
plt.title('Boxplot of Age (Outliers Handled)')
plt.xlabel('Age')
plt.show()
```

#### Min-max scaling

```python
from sklearn.preprocessing import MinMaxScaler

# Extracting the target column 'Survived'
target_column = df['Survived']

# Extracting the features (excluding 'Survived' column)
features = df.drop(columns=['Survived'])

# Assuming 'features' DataFrame contains both numeric and categorical columns
# Extracting categorical columns for encoding
categorical_columns = features.select_dtypes(include=['object']).columns

# One-hot encoding categorical columns
encoded_features = pd.get_dummies(features, columns=categorical_columns)

# Scaling the encoded features using MinMaxScaler for numerical columns
scaler = MinMaxScaler()
scaled_numeric_features = scaler.fit_transform(encoded_features.select_dtypes(include=['float64', 'int64']))

# Creating a DataFrame with scaled features and 'Survived' column
scaled_df = pd.DataFrame(scaled_numeric_features, columns=encoded_features.select_dtypes(include=['float64', 'int64']).columns)
scaled_df['Survived'] = target_column

# Displaying the scaled DataFrame
print(scaled_df)
```
